```python
import pandas as pd

data=pd.read_csv('/content/Training_of_Jail_Officers.CSV')
data
```

```python
import seaborn as sns
a=data[(data['TimePeriod']==2001) & (data['Indicator_Name']=='Total Officials Trained') ]
a
```

```python
import matplotlib.pyplot as plt
sns.barplot(x=a['Area_Name'],y=a['Data_Value'],data=a)
plt.title('No. of Officials Trained in 2001')
plt.xlabel('State')
plt.ylabel('Number')
plt.xticks(rotation='vertical')
plt.show()
```

```python
b=data[(data['Area_Name']=='Delhi') & (data['Indicator_Name']=='Total Officials Trained') ]
b
```

```python
sns.lineplot(x=b['TimePeriod'],y=b['Data_Value'],data=b)
plt.title('No. of Officials Trained in Delhi')
plt.xlabel('Year')
plt.ylabel('Number')
#plt.xticks(rotation='vertical')
plt.show()
```

```python
mental=pd.read_csv('/content/Inmates_suffering_from_mental_ilness.csv')
mental
```

```python
c=mental[mental['YEAR']==2012]
c
```

```python
sns.barplot(x=c['STATE/UT'],y=c['Grand Total'],data=c)
plt.title('Inmates suffering from mental illness in 2012')
plt.xlabel('State_UT')
plt.ylabel('Total')
plt.xticks(rotation='vertical')
plt.show()
```

```python
d=mental[mental['STATE/UT']=='ORISSA']
d
```

```python
sns.lineplot(x=d['YEAR'],y=d['Grand Total'],data=d)
plt.title('Mentally Ill patient trend in ORISSA')
plt.xlabel('Year')
plt.ylabel('Number')
#plt.xticks(rotation='vertical')
plt.show()
```

```python
rec=pd.read_csv('/content/Recidivism.csv')
rec
```
